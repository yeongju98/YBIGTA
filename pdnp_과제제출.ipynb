#### 중요하다고 생각되는 부분만 추출

```python
data2 = data[['quality','residual sugar','pH','alcohol','type']]
data2.head()
```

#### groupby를 통해 quality로 묶고 통계값(min,mean,max) 확인

```python
data2.groupby('quality').describe()
```

#### groupby를 통해 quality/type로 묶고 개수 세기

```python
data2.groupby(['quality','type']).size()
```

#### 알코올 함량을 기준으로 오름차순 정리 (20개)

```python
data2.sort_values('alcohol').head(20)
```

#### 알코올 함량을 기준으로 내림차순 정리 (20개)

```python
data2.sort_values('alcohol', ascending=False).head(20)
```

#### quailty가 7 이상인 데이터 추출 (20개)

```python
data2[data2['quality']>=7].head(20)
```

#### alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (lambda 쓰세요)

```python
data2['alcohol_intensity'] = data2['alcohol'].apply(lambda x: 'strong' if x>10 else 'weak')
data2.head()
```

#### merge 실습을 위해 데이터를 인위적으로 나눠볼게요

```python
data_div1 = data[['quality','fixed acidity','citric acid']].iloc[:5]
data_div1
```

```python
data_div2 = data[['quality','volatile acidity','chlorides','alcohol']].iloc[[11, 12, 17, 25, 45]]
data_div2
```

#### quality를 기준으로 inner join

```python
pd.merge(data_div1, data_div2, how='inner', on='quality')
```

#### quality를 기준으로 outer join

```python
pd.merge(data_div1, data_div2, how='outer', on='quality')
```
